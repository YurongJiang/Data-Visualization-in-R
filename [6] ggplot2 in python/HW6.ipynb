```python
#1 
try:
    n = int(input("Enter a positive integer:"))
    if n >0:
        for i in range(0,n):
            i = i ** 2
            print(i)
    else:
        print("Please a positive integer")
except:
        print("Please a positive integer")
```

```python
#2 
score = float(input("Enter a number:"))

if score > 0 and score <100:
    if score >= 90:
        print("A")
    elif score >= 80:
        print("B")
    elif score >= 70:
        print("C")
    elif score >= 60:
        print("D")
    else:
        print("F")
else:
    print("Bad Score")
```

```python
#3
def leap_year(n):
    if n%4 == 0:
        if n%100 != 0:
            return True
        else:
            if n%400 == 0:              
                return True
            else:
                return False
    else:
        return True

print(leap_year(int(input("Enter a year:"))))
```
